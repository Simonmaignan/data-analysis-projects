## Data Gathering
Let's read the different Excel sheets and load them into data frames

```python
import pandas as pd
```

```python
azure_dk_df = pd.read_csv("dataset/Azure DK.csv")
azure_dk_df.head()
```

```python
azure_es_df = pd.read_csv("dataset/Azure ES.csv")
azure_es_df.head()
```

```python
azure_se_df = pd.read_csv("dataset/Azure SE.csv")
azure_se_df.head()
```

```python
azure_uk_df = pd.read_csv("dataset/Azure UK.csv")
azure_uk_df.head()
```

```python
azure_uk_df.describe()
```

```python
azure_uk_df.info()
```

```python
fx_rates_nov23_df = pd.read_csv("dataset/fx_rates_nov_23.csv")
fx_rates_nov23_df.head()
```

```python
fx_rates_oct23_df = pd.read_csv("dataset/fx_rates_oct_23.csv")
fx_rates_oct23_df.head()
```

```python
fx_rates_oct23_df.info()
```

```python
plataforma_client_xlsx = pd.ExcelFile("dataset/Plataforma Cliente.xlsx")
plataforma_client_df = pd.read_excel(plataforma_client_xlsx)
plataforma_client_df.head()
```

```python
azure_dk_df["Usage Date"] = pd.to_datetime(azure_dk_df["Usage Date"])
azure_es_df["Usage Date"] = pd.to_datetime(azure_es_df["Usage Date"])
azure_se_df["Usage Date"] = pd.to_datetime(azure_se_df["Usage Date"])
azure_uk_df["Usage Date"] = pd.to_datetime(azure_uk_df["Usage Date"])
fx_rates_oct23_df["Date"] = pd.to_datetime(fx_rates_oct23_df["Month"])
fx_rates_nov23_df["Date"] = pd.to_datetime(fx_rates_nov23_df["Month"])
```

```python
azure_dk_df.describe()
```

```python
azure_dk_df["Usage Month"] = azure_dk_df["Usage Date"].dt.month
azure_dk_df["Usage Year"] = azure_dk_df["Usage Date"].dt.year
azure_es_df["Usage Month"] = azure_es_df["Usage Date"].dt.month
azure_es_df["Usage Year"] = azure_es_df["Usage Date"].dt.year
azure_se_df["Usage Month"] = azure_se_df["Usage Date"].dt.month
azure_se_df["Usage Year"] = azure_se_df["Usage Date"].dt.year
azure_uk_df["Usage Month"] = azure_uk_df["Usage Date"].dt.month
azure_uk_df["Usage Year"] = azure_uk_df["Usage Date"].dt.year
fx_rates_nov23_df["Month"] = fx_rates_nov23_df["Date"].dt.month
fx_rates_nov23_df["Year"] = fx_rates_nov23_df["Date"].dt.year
fx_rates_oct23_df["Month"] = fx_rates_oct23_df["Date"].dt.month
fx_rates_oct23_df["Year"] = fx_rates_oct23_df["Date"].dt.year
```

```python
azure_merged_df = pd.concat(
    [azure_dk_df, azure_es_df, azure_se_df, azure_uk_df]
)
azure_merged_df.info()
```

```python
azure_merged_df = pd.merge(
    azure_merged_df,
    plataforma_client_df,
    how="left",
    left_on="API Country",
    right_on="Country",
)
azure_merged_df.drop(["Customer", "Country"], axis=1, inplace=True)
azure_merged_df.head()
```

```python
fx_rates_df = pd.concat([fx_rates_nov23_df, fx_rates_oct23_df])
fx_rates_df.drop("Date", axis=1, inplace=True)
fx_rates_df.head()
```

```python
fx_rates_df.info()
```

```python
azure_merged_df = pd.merge(
    azure_merged_df,
    fx_rates_df,
    how="left",
    left_on=["Usage Month", "Usage Year", "API Currancy"],
    right_on=["Month", "Year", "CurrencyCode"],
)
azure_merged_df.drop(
    ["CurrencyCode", "Month", "Year", "Usage Month", "Usage Year"],
    axis=1,
    inplace=True,
)
azure_merged_df.info()
```

```python
azure_merged_df.head()
```

```python
azure_merged_df.describe()
```

```python
azure_merged_df["Billing Pre Tax Total"] = (
    azure_merged_df["Billing Pre Tax Total"] * azure_merged_df["USDPerUnit"]
)
azure_merged_df.describe()
```

```python
azure_merged_df.to_csv("dataset/azure_merged.csv", index=False)
```
